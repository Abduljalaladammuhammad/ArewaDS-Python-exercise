## EXERCISE TWO

```python
sum = 0
for i in range(100):
    sum += i
    print(sum)
```

```python
sum_even = 0
sum_odd = 0
for i in range(100):
    if i % 2 ==0:
        sum_even += i
    else:
         sum_odd += i
   
    
    print("the sum of all even is", sum_even )
    print("the sum of all odd is", sum_odd)
```
